### ðŸ“– Concept: What is a DataFrame?

A DataFrame is a 2-dimensional labeled data structure with columns (like a spreadsheet or SQL table).
- **Rows**: Individual records (e.g., each campaign, each day, each customer)
- **Columns**: Variables/features (e.g., cost, conversions, channel)
- **Index**: Row labels (by default, numbers 0, 1, 2...)

### ðŸ“– Concept: Boolean Filtering

Filter rows that meet specific conditions:

```python
# Recreate our campaigns DataFrame with metrics
df_campaigns = pd.DataFrame(campaigns)
df_campaigns['roas'] = df_campaigns['revenue'] / df_campaigns['cost']
df_campaigns['cpa'] = df_campaigns['cost'] / df_campaigns['conversions']

# Filter: Campaigns with ROAS >= 3.0
high_roas = df_campaigns[df_campaigns['roas'] >= 3.0]
print("High ROAS Campaigns (>= 3.0x):")
print(high_roas[['campaign_name', 'channel', 'roas']])

# Filter: Multiple conditions with &
good_performers = df_campaigns[(df_campaigns['roas'] >= 3.0) & (df_campaigns['conversions'] >= 300)]
print("\nGood Performers (ROAS >= 3.0 AND Conversions >= 300):")
print(good_performers[['campaign_name', 'roas', 'conversions']])

# Filter: OR condition with |
attention_needed = df_campaigns[(df_campaigns['roas'] < 2.0) | (df_campaigns['cpa'] > 20)]
print("\nCampaigns Needing Attention (Low ROAS OR High CPA):")
print(attention_needed[['campaign_name', 'roas', 'cpa']])
```

### ðŸ“– Concept: Sorting Data

Sort by one or more columns:

### ðŸ“– Concept: loc vs iloc

## ðŸ“… Day 10: GroupBy and Aggregations (~60 min)

### Learning Objectives
- Group data by categories
- Perform aggregations (sum, mean, count, etc.)
- Create pivot tables
- Calculate metrics by segment

### The Business Problem
You need to analyze performance by channel, campaign type, date, or other dimensions. GroupBy is your most powerful tool for this.

### ðŸ“– Concept: Basic GroupBy

The GroupBy pattern: Split â†’ Apply â†’ Combine

```python
# Group by channel and sum
channel_totals = df_campaigns.groupby('channel')[['cost', 'revenue', 'conversions']].sum()
print("Performance by Channel:")
print(channel_totals)

# Calculate blended ROAS by channel
channel_totals['blended_roas'] = channel_totals['revenue'] / channel_totals['cost']
print("\nWith Blended ROAS:")
print(channel_totals)
```

### ðŸ“– Concept: Multiple Aggregations

Apply different functions to different columns:

```python
# Multiple aggregations at once
channel_summary = df_campaigns.groupby('channel').agg({
    'campaign_name': 'count',  # Number of campaigns
    'cost': 'sum',
    'revenue': 'sum',
    'conversions': 'sum',
    'roas': 'mean'  # Average ROAS
})

channel_summary.columns = ['num_campaigns', 'total_cost', 'total_revenue', 'total_conversions', 'avg_roas']
channel_summary['blended_roas'] = channel_summary['total_revenue'] / channel_summary['total_cost']

print("Comprehensive Channel Analysis:")
print(channel_summary)
```

### ðŸ“– Concept: Time-Based Data

Let's work with daily campaign data:

```python
# Create daily campaign data
daily_data = {
    'date': pd.date_range('2024-01-01', periods=30, freq='D').tolist() * 3,
    'campaign': ['Search Brand'] * 30 + ['Facebook Prospecting'] * 30 + ['Display Remarketing'] * 30,
    'cost': np.random.uniform(100, 500, 90),
    'conversions': np.random.randint(5, 30, 90),
    'revenue': np.random.uniform(300, 1500, 90)
}

df_daily = pd.DataFrame(daily_data)
df_daily['roas'] = df_daily['revenue'] / df_daily['cost']

print("Daily Campaign Data (first 10 rows):")
print(df_daily.head(10))
```

```python
# Weekly aggregation
df_daily['week'] = df_daily['date'].dt.to_period('W')
weekly_summary = df_daily.groupby(['week', 'campaign']).agg({
    'cost': 'sum',
    'conversions': 'sum',
    'revenue': 'sum'
}).reset_index()

weekly_summary['weekly_roas'] = weekly_summary['revenue'] / weekly_summary['cost']

print("\nWeekly Performance by Campaign:")
print(weekly_summary.head(12))
```
